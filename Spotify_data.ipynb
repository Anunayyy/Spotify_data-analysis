```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## data from 1922- 2021

### here we have used a lot of different atrributes to distinguish songs from one another

```python
df_tracks = pd.read_csv('D:/spotify/tracks.csv')
df_tracks.head()
```

## now we check the data set

```python
# null values

pd.isnull(df_tracks).sum()
```

```python
## information about the dataset
df_tracks.info()
```

```python
## least popular songs
sorted_df = df_tracks.sort_values('popularity', ascending= True).head(10)
sorted_df
```

```python
## most popular songs
sorted_df_ = df_tracks.sort_values('popularity', ascending= True).tail(10)
sorted_df_
```

```python
## describing the dataset
df_tracks.describe().transpose()
```

```python
## 10 most popular songs
most_popular = df_tracks.query('popularity>90', inplace = False).sort_values('popularity',ascending=False)  ## here we used inplace because we dont want to change our data set. and for descending we can use ascending= false
most_popular[:10]
```

```python
## data is sorted on the bais of release date
df_tracks.set_index("release_date",inplace = True)
df_tracks.index=pd.to_datetime(df_tracks.index)
df_tracks.head()
```

```python
## to find any specific information
## here we try to find the artist in the 33rd column
df_tracks[["artists"]].iloc[33]  ##iloc = index location
```

```python
## to convert miliseconds into seconds
df_tracks["duration"]= df_tracks['duration_ms'].apply(lambda x: round(x/1000))
df_tracks.drop("duration_ms", inplace=True, axis=1)
```

```python
df_tracks.duration.head()
```

## co-relation map

```python
corr_df = df_tracks.drop(["key","mode","explicit"],axis=1).corr(method='pearson')
plt.figure(figsize=(14,6))
heatmap = sns.heatmap(corr_df,annot=True,fmt=".1g",vmin=-1, center=0, cmap="inferno", linewidth=1, linecolor="Black")
heatmap.set_title("Correlation HeatMap Between Variable")
heatmap.set_xticklabels(heatmap.get_xticklabels(), rotation=90)
```

```python
## to find.4% of the total dataset
sample_df = df_tracks.sample(int(0.004*len(df_tracks)))
print(len(sample_df))
```

## plotting a correlation graph between loudness and energy

```python
plt.figure(figsize=(20,10))
sns.regplot(data = sample_df, y = "loudness", x = "energy", color = "c").set(title="loudness vs energy")
```

## correlation between popularity and dancebility

```python
plt.figure(figsize=(10,6))
sns.regplot(data = sample_df, y = "popularity", x = "danceability", color = "c").set(title="popularity and dancebility")
```

# using this we can find out the correlation between different attributes defined in a song and come to certain conclusions

```python
df_tracks['dates']=df_tracks.index.get_level_values('release_date')
df_tracks.dates=pd.to_datetime(df_tracks.dates)
years=df_tracks.dates.dt.year
```

# bar plot on year vs duration

```python
total_dr = df_tracks.duration
fig_dims = (18,7)
fig, ax= plt.subplots(figsize = fig_dims)
fig = sns.barplot(x=years, y=total_dr, ax= ax, errwidth=False).set(title="year vs Duration")
plt.xticks(rotation=90)
```

# year vs duration line plot

```python
total_dr=df_tracks.duration
sns.set_style(style="whitegrid")
figs_dims = (10,5)
fig, ax = plt.subplots(figsize=fig_dims)
fig=sns.lineplot(x=years,y=total_dr,ax=ax).set(title="year vs duration")
plt.xticks(rotation=90)
```
